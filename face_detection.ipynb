```python
from deepface import DeepFace

from pathlib import Path

import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Conference https://www.youtube.com/watch?v=MYNMsC8The0
```

```python
data_path = Path("./data/faces")

# Load images of two different persons 
image_1 = Image.open(data_path / "profile1.jpg")
image_2 = Image.open(data_path / "profile2.jpg")

fig, ax = plt.subplots(ncols=2, figsize=(10, 4))

ax[0].imshow(image_1)
ax[1].imshow(image_2)
```

```python
result = DeepFace.verify(
    img1_path = data_path / "profile1.jpg",
    img2_path = data_path / "profile2.jpg",
    model_name="Facenet",
    enforce_detection=False,
)

result
```

```python
result = DeepFace.verify(
    img1_path = data_path / "profile1.jpg",
    img2_path = data_path / "profile2.jpg",
    model_name="OpenFace",
    enforce_detection=False,
    threshold=0.1
)

result
```
